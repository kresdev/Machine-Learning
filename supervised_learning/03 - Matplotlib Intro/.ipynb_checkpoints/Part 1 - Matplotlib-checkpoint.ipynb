```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

## Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1)
ax.plot(X, y2)
```

### shortcut

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r-.', linewidth=5)
plt.plot(X, y2, 'bo', markersize=3)
plt.ylim(-2, 2)
plt.xlim(0, 10)
plt.axis('equal')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sin dan Cos')
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r-.', linewidth=5)
ax.plot(X, y2, 'bo', markersize=3)
ax.set(xlim=(0, 10), ylim=(-2, 2))
ax.set_title('Sin dan Cos')
ax.set_xlabel('X')
ax.set_ylabel('y')
ax.axis('equal');
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(X, y1, 'r-.', linewidth=5, label='Sin')
ax.plot(X, y2, 'bo', markersize=3, label='Cos')
ax.set(xlim=(0, 10), ylim=(-2, 2))
ax.set_title('Sin dan Cos')
ax.set_xlabel('X')
ax.set_ylabel('y')
ax.axis('equal');
ax.legend(loc='lower center', ncol=2, shadow=True)
```

## Scatter Plot

```python
df = pd.read_csv('data/circle_data.csv', delimiter=';')
df.head()
```

```python
X = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(X,y)
plt.axis('equal')
```

```python
plt.scatter(X,y, c=rad , s=rad*90, cmap='seismic')
plt.axis('equal')
```
